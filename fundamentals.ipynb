# Creating arrays

```python
# 0 D array
import numpy as np
a =np.array(42)
print(a)
```

```python
# 1 D array

# using list
import numpy as np
a = np.array([1,2,3,4,5])
print(a)

# using tuple
import numpy as np
a = np.array((1,2,3,4,5))
print(a)
```

```python
# 2 D array
import numpy as np
b = np.array([[1,2,3],[4,5,6]])
print("2 Dim array")
print(b)

# 3 D array
c = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
print("\n3 Dim array")
print("\n",c)
```

```python
print("No. of Dimensions =",a.ndim)
print("No. of Dimensions =",b.ndim)
print("No. of Dimensions =",c.ndim)
```

```python
# Creation of arrays with different dimensions
d = np.array(1,ndmin=5)
print(d)
d = np.array([[1,2,3],[1,2,3]],ndmin=5)
print(d)
```

# Array indexing

```python
# 1 d
a = np.array([1,2,3,4,5])
print(a[1])
```

```python
# 2 d
b = np.array([[1,2,3],[4,5,6]])
print(b[0],b[1]) # accessing all elements in each dimension
print(b[0,1], b[1,0]) # accessing elements individually in each dimension

# 3 d
c = np.array([[[1,2,3],[4,5,6]],[[11,12,13],[14,15,16]]])
print(c[0,1,2]) # 0 => [1,2,3],[4,5,6]. 1 => [4,5,6]. 2 => 6
```
